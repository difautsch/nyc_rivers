```python
import pandas as pd
f = open('2019_raw.csv','r')
data = f.read()
rows = data.split('\n')
test = []

pd.set_option('display.max_rows', 1000)
#three lists: site_data and site_name and year
site_data = []
site_names = []
year = []
#data is most definitely 2018_raw
#print (data)
for i in rows:
    i = i.split(':')
    test.append(i)
#list element of test corresponds to a site's testing values    
#print (test[4:7])
for item in test[4:]:
    #within 'item' except for first item, the data is in the second element of the list, and is the second sub-element
    item_split = item[0].split(',')
    site = (item_split[0])
    for items in item[1:]:
        data = items.split(',')
        #two sites causing errors bc blank values and items with only one value
        if len(data) == 1:
            continue
        site_data.append(data[1])
        site_names.append(site)
        year.append('2018')
#convert strings to numeric int32 
for idx, value in enumerate(site_data):
    if value == '<10':
        value = 10
        site_data[idx] = value
    if value == '>24196':
        value = 24196
        site_data[idx] = value
#delete three blank values for error handling
for idx, value in enumerate(site_data):
    try: 
        int (value)
    except:
        del site_data[idx]
        del site_names[idx]
        del year[idx]
        
#create dataframe 
df = pd.DataFrame(list(zip(site_names,site_data,year)), columns = ['site_names','bacteria','year'])
df = df.astype({'bacteria':'int32'})

df.describe()
print (df)
df.to_csv('2019clean')
```
